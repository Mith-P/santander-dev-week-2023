```python
import pandas as pd
import google.generativeai as genai
import time

# 1. Configuração com Modelo Estável
GOOGLE_API_KEY = "AIzaSyDtyfMc8zEfhXwArLK3iKhLBheCh6ZEgB8"
genai.configure(api_key=GOOGLE_API_KEY)

def gerar_mensagem_ia(nome, interesse):
    # Trocando para o modelo 1.0 Pro, que costuma aceitar melhor chaves gratuitas
    model = genai.GenerativeModel('gemini-2.5-flash')

    prompt = f"Olá {nome}, escreva uma frase curta incentivando seus estudos em {interesse}."

    try:
        # Adicionamos um tempo de espera maior para evitar bloqueio por IP
        time.sleep(2)
        response = model.generate_content(prompt)

        # Verificamos se a resposta existe e se não foi bloqueada
        if response and response.candidates:
            # O .text pode falhar se a IA bloquear o conteúdo, tratamos isso aqui
            try:
                return response.text.strip()
            except ValueError:
                return "Conteúdo bloqueado pelos filtros da IA."
        return "Resposta vazia da IA."

    except Exception as e:
        # Se der erro 400 aqui, ele vai imprimir a mensagem exata do Google
        print(f"Erro detalhado para {nome}: {e}")
        return "Erro na geração"

## --- E e L (Extract e Load) ---

def transformar_e_carregar():
    print("--- [E] Extraindo ---")
    df = pd.read_csv('clientes.csv')

    print("--- [T] Transformando (IA) ---")
    mensagens = []
    for index, row in df.iterrows():
        print(f"Processando: {row['Nome']}...")
        msg = gerar_mensagem_ia(row['Nome'], row['Interesse'])
        mensagens.append(msg)

    df['Mensagem'] = mensagens

    print("--- [L] Carregando ---")
    df.to_csv('clientes_finalizados.csv', index=False, encoding='utf-8-sig')
    print("Processo concluído! Verifique o arquivo 'clientes_finalizados.csv'.")

# Executar
if __name__ == "__main__":
    transformar_e_carregar()
```
